```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
import os
import pandas as pd
from selenium import webdriver
```

```python
column_names = ['date','away_team','away_1st_inn','home_team','home_1st_inn']

total_df = pd.DataFrame(columns = column_names)
total_df
```

```python
# url = 'https://www.baseball-reference.com/boxes/'

url = 'https://www.baseball-reference.com/boxes/?year=2021&month=4&day=1'

response = requests.get(url)

soup = bs(response.text, 'html.parser')
```

```python
#COULD BE USED TO FIND A SPECIFIC HTML TAG TO RETURN
results = soup.find_all('td', class_='right gamelink')
print(results)
```

```python
results2 = str(results)
soup2 = bs(results2)
# FOR LOOP TO FIND SPECIFIC ENDPOINT FOR EACH GAME'S SCORES
for a in soup2.find_all('a', href=True):
    print("Found the URL:", a['href'])
```

```python
# FOR LOOP TO CREATE EACH SPECIFIC LINK TO EACH GAME'S SCORES
url_list = []
for a in soup2.find_all('a', href=True):
    link = 'https://www.baseball-reference.com' + a['href']
    url_list.append(link)
    print(link)
```

```python
# LOOPING THROUGH EACH GAME URL TO PULL BOX INFO
# date = soup.find_all('div', class_="scorebox_meta")[0]
# dt = 1
# for d in date:
#     if dt == 2:
# #     if dt == 2 or dt == 3:
#         print(d.text)
#     dt+=1
game_date_list = []
away_team_list = []
away_score_list = []
home_team_list = []
home_score_list = []
for url in url_list:
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    
    date = soup.find_all('div', class_="scorebox_meta")[0]
    dt = 1
    for d in date:
        if dt == 2:
            final_date = d.text
            print(final_date)
            game_date_list.append(final_date)
#             print(d.text)

        dt+=1
    #LOCATES A SPECIFC 'TR'. SET INDEXES TO ONLY PULL TEAM ROWS    
    final_tr1 = soup.find_all('tr')[1]
    final_tr2 = soup.find_all('tr')[2]
    y = 1
    for r in final_tr1:
        if y == 4:
            away_team = r.text
            print(away_team)
            away_team_list.append(away_team)
        elif y == 6:
            away_tr = r.text
            print(away_tr)
            away_score_list.append(away_tr)
            
        y+=1
    y = 1    
    for r in final_tr2:
        if y == 4:
            home_team = r.text
            print(home_team)
            home_team_list.append(home_team)
        if y == 6:
            home_tr = r.text
            print(home_tr)
            home_score_list.append(home_tr)
        y+=1
```

```python
dict = {'date': game_date_list,
        'away_team': away_team_list,
       'away_1st_inn': away_score_list,
       'home_team': home_team_list,
       'home_1st_inn': home_score_list}
```

```python
df = pd.DataFrame(dict)
df
```

```python
total_df = total_df.append(df, ignore_index = True)
total_df.tail(50)
```
