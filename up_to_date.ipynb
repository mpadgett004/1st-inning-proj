```python
results = soup.find_all('td', class_='right gamelink')
string_results = str(results)
soup = bs(string_results)
# print(results)
# FOR LOOP TO CREATE EACH SPECIFIC LINK TO EACH GAME'S SCORES
url_list = []
for a in soup.find_all('a', href=True):
    link = 'https://www.baseball-reference.com' + a['href']
    url_list.append(link)
    print(link)
```

```python
game_date_list = []
away_team_list = []
away_score_list = []
home_team_list = []
home_score_list = []
for url in url_list:
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    
    date = soup.find_all('div', class_="scorebox_meta")[0]
    dt = 1
    for d in date:
        if dt == 2:
            final_date = d.text
            print(final_date)
            game_date_list.append(final_date)
#             print(d.text)

        dt+=1
    #LOCATES A SPECIFC 'TR'. SET INDEXES TO ONLY PULL TEAM ROWS    
    final_tr1 = soup.find_all('tr')[1]
    final_tr2 = soup.find_all('tr')[2]
    y = 1
    for r in final_tr1:
        if y == 4:
            away_team = r.text
            print(away_team)
            away_team_list.append(away_team)
        elif y == 6:
            away_tr = r.text
            print(away_tr)
            away_score_list.append(away_tr)
            
        y+=1
    y = 1    
    for r in final_tr2:
        if y == 4:
            home_team = r.text
            print(home_team)
            home_team_list.append(home_team)
        if y == 6:
            home_tr = r.text
            print(home_tr)
            home_score_list.append(home_tr)
        y+=1
```

```python
dict = {'date': game_date_list,
        'away_team': away_team_list,
       'away_1st_inn': away_score_list,
       'home_team': home_team_list,
       'home_1st_inn': home_score_list}
```

```python
df = pd.DataFrame(dict)
df
```

```python
updated_df = full_df.append(df, ignore_index = True)
updated_df.tail(25)
```

```python
updated_df.to_csv('full_firstinning_data.csv', index=False)
```
